```python
paragraphs = []

with open("all_paragraphs_summarized.txt", 'r') as file:
    paragraphs = file.read().split('\n')
# Optionally, to remove any empty strings if there are extra newlines
paragraphs = [para for para in paragraphs if para]
```

```python
from openai import OpenAI

client = OpenAI(api_key = YOUR_API_KEY)

def generate_qa_pairs(paragraph):
    try:
        response = client.chat.completions.create(
          model="gpt-3.5-turbo",
          messages=[
              {"role": "system", "content": "You are a helpful assistant."},
              {"role": "user", "content": f"Directly generate specific questions targeting key information in the following text segment, along with concise answers directly extracted from the text. Format each question-answer pair as follows: 'Q: [Question] \n A: [Answer]'.\n\n{paragraph}"}
          ]
        )
        return response.choices[0].message.content
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Replace this with your list of sample paragraphs
qa_pairs_list = []
for paragraph in paragraphs:
    qa_pairs = generate_qa_pairs(paragraph)
    print(f"Paragraph: {paragraph}\nGenerated Questions:\n{qa_pairs}\n")
    # Assuming each 'qa_pairs' is a string formatted as "Q: Question / A: Answer"
    # Splitting it into a list of Q/A pairs
    qa_pairs_list.append(qa_pairs)


with open("question_answer_pairs.txt", 'w') as file:
    for qa_pair in qa_pairs_list:
        file.write(qa_pair + '\n')
```
